# Using Matplotlib and Seaborn on UCI machine learning red wine data set for visulizations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
red_wine_url = "https://raw.githubusercontent.com/nishilp/DataScience01/master/Data/red_wine.csv"
df = pd.read_csv(red_wine_url)
```

```python
df.head()
```

```python
sum(df.duplicated())
```

```python
df = df.drop_duplicates()
```

```python
area1 = df[df['good'] == 1]
area1 = area1['good']
area1.head()
```

```python
area2 = df[df['good'] == 0]
area2 = area2['good']
area2.head()
```

```python
colors = df['good']
```

# Matplotlib

```python
fig, ax = plt.subplots()
ax.scatter(df['fixed acidity'], df['volatile acidity'], s=area1, c=colors, marker='^') #c=c)
ax.scatter(df['fixed acidity'], df['volatile acidity'], s=area2, c=colors, marker='o') #c=c)
ax.set_xlabel(r'Fixed Acidity', fontsize=15)
ax.set_ylabel(r'Volatile Acidity', fontsize=15)
plt.show()
```
